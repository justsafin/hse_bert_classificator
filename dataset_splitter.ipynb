```python
def get_count_dict(df, column_name) -> dict:
    flattened_list = [item for sublist in df[column_name] for item in sublist]
    count_dict = dict(Counter(flattened_list))
    return count_dict
```

```python
def get_drop_cats_list(count_dict, treshold=1):
    dropped_cats = []
    potential_drop = {}
    for key, value in list(count_dict.items()):
        if value <= treshold:
            dropped_cats.append(key)
            del count_dict[key]
        elif value <= 25:
            # print(key, value)
            potential_drop[key] = value
    # print(f"dropped_cats = {dropped_cats}")
    # print(f"potential_drop = {potential_drop}")
    return dropped_cats, potential_drop
```

```python
def drop_cats_from_df(df, cats_to_drop, level='RGNTI_L1'):
    
    def remove_values(lst, values_to_remove):
        return [item for item in lst if item not in values_to_remove]

    screen = set(cats_to_drop)
    
    print(df.shape)

    df[level] = df[level].apply(lambda x: remove_values(x, screen))
    df = df[df[level].map(len) > 0]
    print(df.shape)
    return df
```

```python
def is_possible_to_add_to_df(row_cats, cats_dict_counter):
    for cat in row_cats:
        num = cats_dict_counter.get(cat, None)
        if num == 0:
            return False
        elif num is None:
            # print(f"undefined category {cat}")
            raise Exception(f"undefined category {cat}")
    return True
```

```python
def descrese_count(cat_list, cat_dict):
    for cat in cat_list:
        if cat in cat_dict:
            cat_dict[cat] -= 1
    return cat_dict
```

```python
from math import ceil

teach_data = {}
test_data = {}

for key, value in list(count_dict_l2.items()):
    test_count = value//2
    test_count = ceil(value * 0.20)
    test_data[key] = test_count
    
    teach_count = value - test_count
    teach_data[key] = teach_count

# проверка
for key in count_dict_l2.keys():
    assert teach_data[key] + test_data[key] == count_dict_l2[key]
```

```python
def sort_dataframe(df, level='RGNTI_L1'):
    df['list_length'] = df[level].apply(len)  # Create a temporary column for list length
    df = df.sort_values(by='list_length', ascending=False)  # Sort by list length in descending order
    df = df.drop(columns='list_length')

    df = df.reset_index(drop=True)
    return df
```

```python
def split_dataframe(df, teach_data, test_data, level="RGNTI_L1"):
    df_test = pd.DataFrame(columns=['data', 'RGNTI_L1', 'RGNTI_L2'])
    df_teach = pd.DataFrame(columns=['data', 'RGNTI_L1', 'RGNTI_L2'])
    unsorted = pd.DataFrame(columns=['data', 'RGNTI_L1', 'RGNTI_L2'])
    for ind in df.index:
        # print(df['RGNTI_L1'][ind])
        try:
            if is_possible_to_add_to_df(df[level][ind], test_data):
                # print("adding to test")
                df_test = df_test._append([df.iloc[ind]])
                test_data = descrese_count(df[level][ind], test_data)
            elif is_possible_to_add_to_df(df[level][ind], teach_data):
                # print("adding to teach")
                df_teach = df_teach._append([df.iloc[ind]])
                teach_data = descrese_count(df[level][ind], teach_data)
            else:
                # print("adding to unsorted")
                unsorted._append([df.iloc[ind]])

        except Exception as e:
            print(f"Found exception {e} \n On row {ind}")
            continue
    return df_teach, df_test, unsorted
```

```python
def create_l1(list_l2):
    x = set()
    for key in list_l2:
        l1_key = key.split(".")[0]
        x.add(l1_key)
    return list(x)
```

```python
def get_dataset(path_to_csv: str, level: str): 
    df = pd.read_csv(path_to_csv, sep="\t") 
    df["labels"] = df[level].apply(literal_eval) 
 
    mlb = MultiLabelBinarizer() 
    sparce_labels = mlb.fit_transform(df["labels"]) 
    classes = mlb.classes_ 
    ready_dataset = sparce_labels 
 
    return ready_dataset, classes
```

```python
train_data = "teach_slice_80_l2_drop_wasted_v2.csv" 
test_data = "test_slice_20_l2_drop_wasted_v2.csv" 
level = "RGNTI_L2" 
 
# Calculate classes 
print("Read train data") 
train_dataset, train_classes = get_dataset(train_data, level) 
test_dataset, test_classes = get_dataset(test_data, level) 
 
ratio = train_dataset.sum(axis=0) / test_dataset.sum(axis=0) 
print(train_dataset.sum(axis=0)) 
print(ratio) 
print(max(ratio)) 
print(min(ratio))
```
